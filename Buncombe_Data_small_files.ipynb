```python
# Import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Read in the two voter_hist.csvs 
voter_hist_df1 = pd.read_csv("voter_hist1.csv")
voter_hist_df2 = pd.read_csv("voter_hist2.csv")
```

```python
# Combine voter_hist csvs into one
voter_hist_large = voter_hist_df1.append(voter_hist_df2, ignore_index=True)
voter_hist_large
```

```python
# Create subset dataframe that adheres to Heroku's 10,000 row max
voter_hist = voter_hist_large.iloc[:10000]
voter_hist
```

```python
# Read in two voter_reg.csvs
voter_reg_df1 = pd.read_csv("voter_reg1.csv")
voter_reg_df2 = pd.read_csv("voter_reg2.csv")
```

```python
# Combine two voter_reg csvs into one
voter_reg_large = voter_reg_df1.append(voter_reg_df2, ignore_index=True)
voter_reg_large.head()
```

```python
# Drop unnecessary columns
voter_reg_large.drop(['dist_2_desc', 'dist_1_desc', 'munic_dist_desc', 'rescue_dist_desc', 'sanit_dist_desc', 'sewer_dist_desc',
                  'water_dist_desc', 'fire_dist_desc', 'school_dist_desc', 'township_desc', 'county_commiss_desc', 
                  'ward_desc', 'municipality_desc', 'absent_ind', 'name_prefx_cd', 'dist_2_abbrv'], axis=1, inplace=True)
voter_reg_large
```

```python
# Get list of columns to facilitate division into additonal dataframes
voter_reg_large.columns
```

```python
# Create subset dataframe that adheres to Heroku's 10,000 row max
voter_reg = voter_reg_large.iloc[:10000]
voter_reg
```

```python
# Pull out demographic information columns to create voter demographic dataframe
voter_demo = voter_reg[['res_city_desc', 'state_cd', 'zip_code', 'race_code', 'ethnic_code', 'party_cd', 'gender_code',\
                       'birth_age', 'birth_state', 'birth_year', 'drivers_lic', 'registr_dt', 'status_cd', 'voter_status_desc',\
                       'reason_cd', 'voter_status_reason_desc', 'ncid']]
voter_demo
```
