```python
import pandas as pd
```

```python
# Read in the two voter_hist.csvs 
voter_hist_df1 = pd.read_csv("voter_hist1.csv")
voter_hist_df2 = pd.read_csv("voter_hist2.csv")
```

```python
# Combine voter_hist csvs into one
voter_hist = voter_hist_df1.append(voter_hist_df2, ignore_index=True)
voter_hist
```

```python
# Read in voter_reg.csvs
voter_reg_df1 = pd.read_csv("voter_reg1.csv")
voter_reg_df2 = pd.read_csv("voter_reg2.csv")
```

```python
# Combine two voter_reg csvs into one
voter_reg = voter_reg_df1.append(voter_reg_df2, ignore_index=True)
voter_reg.head()
```

```python
# Drop unnecessary columns
voter_reg.drop(['dist_2_desc', 'dist_1_desc', 'munic_dist_desc', 'rescue_dist_desc', 'sanit_dist_desc', 'sewer_dist_desc',
                  'water_dist_desc', 'fire_dist_desc', 'school_dist_desc', 'township_desc', 'county_commiss_desc', 
                  'ward_desc', 'municipality_desc', 'absent_ind', 'name_prefx_cd', 'dist_2_abbrv'], axis=1, inplace=True)
voter_reg
```

```python
# Get list of columns
voter_reg.columns
```

```python
# Pull out personally identifying information into 
voter_identifiers = voter_reg[['last_name', 'first_name', 'middle_name', 'name_suffix_lbl', 'res_street_address', 'mail_addr1',\
                               'mail_addr2', 'mail_addr3', 'mail_addr4', 'mail_city', 'mail_state', 'mail_zipcode',\
                               'full_phone_number',]]
voter_identifiers.head()
```

```python
# Pull out demographic information columns to create voter demographic dataframe
voter_demo = voter_reg[['res_city_desc', 'state_cd', 'zip_code', 'race_code', 'ethnic_code', 'party_cd', 'gender_code',\
                       'birth_age', 'birth_state', 'birth_year', 'drivers_lic', 'registr_dt', 'status_cd', 'voter_status_desc',\
                       'reason_cd', 'voter_status_reason_desc', 'ncid']]
voter_demo.head()
```

```python
# Pull out jurisdiction and county information columns to create voter dist dataframe
voter_dist = voter_reg[['county_id', 'county_desc', 'precinct_abbrv', 'precinct_desc', 'municipality_abbrv',\
                                'ward_abbrv', 'cong_dist_abbrv', 'super_court_abbrv', 'judic_dist_abbrv', 'nc_senate_abbrv',\
                                'nc_house_abbrv', 'county_commiss_abbrv', 'township_abbrv', 'school_dist_abbrv',\
                                'fire_dist_abbrv', 'water_dist_abbrv', 'sewer_dist_abbrv', 'sanit_dist_abbrv',\
                                'rescue_dist_abbrv', 'munic_dist_abbrv', 'dist_1_abbrv', 'ncid',\
                                'vtd_abbrv', 'vtd_desc\r']]
voter_dist.head()
```

```python
# Read in county codes csv to create dataframe
county_codes = pd.read_csv("county_codes.csv")
county_codes
```

```python
# Create ethnic codes dataframe from information provided on the NC Board of Elections website
ethnic_codes = pd.DataFrame(data={'ethnicity': ['HL', 'NL', 'UN'], 'description': ['Hispanic or Latino', 'Not Hispanic or Not Latino', 'Undesignated']})
ethnic_codes
```

```python
# Create race codes dataframe from information provided on the NC Board of Elections website
race_codes = pd.DataFrame(data={'race': ['A', 'B', 'I', 'M', 'O', 'P', 'U', 'W'], 
                                 'description': ['Asian', 'Black or African American', 'American Indian or Alaska Native',
                                                'Two or more races', 'Other', 'Native Hawaiian or Pacific Islander',
                                                'Undesignated', 'White']})
race_codes
```

```python
# Create status codes dataframe from information provided on the NC Board of Elections website
status_codes = pd.DataFrame(data={'status': ['A', 'D', 'I', 'R', 'S'], 
                                  'description': ['Active', 'Denied', 'Inactive', 'Removed', 'Temporary (applicable to military and overseas)']})
status_codes
```
