```python
# Determine number of row in csv
in_csv = 'Resources/voter_hist.csv'

number_lines = sum(1 for row in (open(in_csv)))
number_lines
```

```python
# Split csv into 2 equal sized csv's so that they can be pushed to github
from __future__ import division

numrows = 536053 #number of rows per csv
count = 0 #keep track of chunks
chunkrows = 100000 #read 100k rows at a time
df = pd.read_csv('Resources/voter_hist.csv', sep='\t',  lineterminator='\n', iterator=True, chunksize=chunkrows) 
for chunk in df: #for each 100k rows
    if count <= numrows/chunkrows: #if threshold has not been reached 
        outname = "voter_hist1.csv"
    else:
        outname = "voter_hist2.csv"
    #append each output to same csv, using no header
    chunk.to_csv(outname, header=True, mode='a', index=None)
    count+=1
```
